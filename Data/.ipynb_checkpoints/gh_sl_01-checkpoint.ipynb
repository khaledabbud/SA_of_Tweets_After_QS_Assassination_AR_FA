```python
def clean_text(row):
    '''
    Takes the mentions and the links out of the text
    Arguments: 
     -text: string
    
    return: 
    - A 3 elements tuple that contains:
      -mentions: a list of strings
      -links: a list of stringd
      -text: a cleaned version of the orriginal text
    '''
    text = row['text']
    mentions = re.findall('@[A-Za-z0-9_]+', text)
    links = re.findall('http[^\s#]+|pic.twitter.com/[^\s#]+', text)
    for m in mentions:
        start = text.find(m)
        end = start + len(m)
        text = text[:start] + text[end:]
    sm = ''
    for m in mentions:
        sm = sm + m  + '\n'
    mentions_str = sm.rstrip()
    row['extracted_mentions'] = mentions_str
    for l in links:
        start = text.find(l)
        end = start + len(l)
        text = text[:start] + text[end:]
    sl = ''
    for l in links:
        sl = sl + l + '\n'
    links_str = sl.rstrip()
    row['text_links'] = links_str
    row['cleaned_text'] = text
    
    return row
```

### Cleaning the text by taking out mentions and links

```python
df['text_links'] = df['text'].apply(clean_text)[1]
```
