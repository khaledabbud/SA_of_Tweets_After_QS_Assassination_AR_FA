## Clssifying the Persian Tweets Without Machine Learning

```python
import pandas as pd
import random
df = pd.read_csv("tweets_sent_fa.csv")
```

```python
df.columns
```

### Droping English and Arabic tweets, the rest are Persian, but I made some mistakes when writing the language when I labeled the training set.

```python
df['language'].unique()
```

```python
df.drop(df[df['language'].isin(["en", 'ar'])].index, inplace=True)
```

```python
df['language'] = 'fa'
```

### Droping the unnecesarry characters and unifying the different veriety of characters. 

```python
def remove_unnecessary(txt):
    #remove extra space
    txt = txt.lower()
    txt = ' '.join(txt.split())
    
    replA = ("أ", "إ")    
    chars_to_drop = ( 'ّ', 'َ', 'ِ', 'ُ', 'ْ', 'ً', 'ٍ', 'ٌ', 'ـ' )
    chars_to_clean = ('…','#', '@', '-', '_')
    txt = txt.replace("ة","ه")
    txt = txt.replace("ي", 'ی')
    for a in replA:
        txt = txt.replace(a, "ا")
    for d in chars_to_drop:
        txt = txt.replace(d,"")
    for c in chars_to_clean:
        txt = txt.replace(c," ")
    return txt.strip()
```

```python
df.loc[:, 'cleaned_text'] = df.loc[:, 'cleaned_text'].apply(remove_unnecessary)
```

### Splitting the tweets to those who where labled by hand and those to be labeled

```python
df_s = df[df['sentiment'].isna()]
df_s.reset_index(drop = True, inplace = True)
df_s.shape
```

##### Drop the irrelevant tweets (labeled 10) and properly label the rest: 'O' for tweets that oppose the assasination, 'S' for tweets that support the assasination and 'N' for neutral tweets. Note that those that are labeled 2 are also neutral

```python
df_t = df.dropna(subset=['sentiment'])
df_t.drop(df_t[df_t['sentiment'] == 10].index, inplace=True)
df_t.loc[df_t['sentiment'] == 2, 'sentiment'] = 'N'
df_t.loc[df_t['sentiment'] == 0, 'sentiment'] = 'N'
df_t.loc[df_t['sentiment'] == 1, 'sentiment'] = 'O'
df_t.loc[df_t['sentiment'] == -1, 'sentiment'] = 'S'
df_t.reset_index(drop = True, inplace = True)
df_t.shape
```

```python
df_t['sentiment'].unique()
```

### Defining Negative and Positive keywords and classifying the tweets base on those keywords

```python
neg_list = ['هلاکت', 'پاسدار ', 'تروریست', 'آزاد', 'ج.ا', 'سقط', 'درک', 'مقدس سازی', 'عرزشی','ارزشی', 'ج ا', 'آبان', 'رژیم', 'دژخیم', 'کتلت', 'دیکتاتور', 'اشغالگر', 'قصاب', 'پویا', 'مرگ بر خامنه','براندازم', 'چیز خر', 'سپاه تروریستی', 'جاعش', 'خجسته', 'جحیم', 'آرامتر', 'آرام تر', 'بگا', 'چلاق', 'عاقا', 'کودک', 'حجامت', 'نعش', 'آخوند', 'شادی عمومی', 'جشن', 'پایکوبی', 'متشکرم', 'متشکریم', 'پر شور', 'ینتفض', 'هلاك', 'به گا', 'شادانه', 'اعتراضات سراسری', 'سرنگون', 'بشاش', 'protest', 'idlib', 'آتش کشید', 'شاد', '۱۵۰۰', 'عن', 'مرگ بر جمهوری', 'غلط', 'بی دفاع']
pos_list = ['سردار', 'سپهبد', 'انتقام سخت', 'شهید', 'شهادت', 'حاج', 'صاحب', 'آقاجان', 'آقا جان', 'مقاومت', 'رهبر انقلاب', 'رهبرانقلاب', 'عشق', 'عاشق', 'ایشان', 'بخون', 'کاخ سفید', 'امام', 'حضرت', 'حجت', 'عاشورا', 'شیفت شب', 'راه خدا', 'نابود', 'تحسبن', 'صهیو', 'بغض', 'حسرت', 'رهبری', 'بیدارتر', 'انتقامی سخت', 'انتقامت',  'جمعه', 'الموت', 'لبیک', 'الفرج', 'لبيك', 'مرگ بر امر', 'مرگ بر آمر', 'اشک رهبر', 'پیکر', 'زینب', 'هزاران قاسم', 'همه قاسم', 'من قاسم', 'علم', 'قلب ما','ماتم کده' ,'سرباز وطن', 'فاطمه', 'مالک', 'شهدا', 'شکوه', 'انقلابی', 'تفنگ پدری', 'سننتقم', 'revenge', 'down', 'روضه', 'نجس', 'آل مروان', 'ترامپ احمق', 'گرگ', 'سازش', 'نهضت', 'ناموس', 'پرواز', 'انتقام می', 'مزدور', 'تشییع', 'سرباز', 'روحت شاد', 'انتقام خو', 'شیر', 'یادش گر', 'لثارات', 'خون تو', 'برکت', 'تنگ', 'حسین', 'الهی بمیر', 'زکیه', 'آسمان', 'لعنت', 'حافظ امنیت', 'آمریکایی', 'خون پدر', 'مثل پدر', 'عطر', 'جهاد', 'سلیمانی هست', 'مدافع', 'آرزو', 'امنیت ایران', 'حصیر', 'قاصم', 'دلگی', 'دوران بزن', 'مرگ بر اسر']
pos_keywords = {}
neg_keywords = {}
```

```python
def process_tweets(row):
    for n in neg_list:
        if n in row:
            if n in neg_keywords:
                neg_keywords[n] += 1
            else:
                neg_keywords[n] = 1
            return 'S'
        
    for p in pos_list:
        if p in row:
            if p in pos_keywords:
                pos_keywords[p] += 1
            else:
                pos_keywords[p] = 1
            return 'O'
    return 'N'
```

```python
df_s.loc[:, 'sentiment'] = df_s.loc[:, 'cleaned_text'].apply(process_tweets)
```

```python
df_pos = df_s[df_s['sentiment'] == 'O']
print("Positive: ", df_pos.shape[0])
df_neg = df_s[df_s['sentiment'] == 'S']
print("Negative: ", df_neg.shape[0])
df_neu = df_s[df_s['sentiment'] == 'N']
print("Neutral: ", df_neu.shape[0])
```
