```python
def remove_unnecessary(txt):
    #remove extra space
    txt = txt.lower()
    txt = ' '.join(txt.split())
    
    replA = ("أ", "إ")    
    chars_to_drop = ( 'ّ', 'َ', 'ِ', 'ُ', 'ْ', 'ً', 'ٍ', 'ٌ', 'ـ' )
    chars_to_clean = ('…','#', '@', '-', '_')
    txt = txt.replace("ة","ه")
    txt = txt.replace("ي", 'ی')
    for a in replA:
        txt = txt.replace(a, "ا")
    for d in chars_to_drop:
        txt = txt.replace(d,"")
    for c in chars_to_clean:
        txt = txt.replace(c," ")
    return txt.strip()
```

```python
conn.create_function("remove_unnecessary", 1, remove_unnecessary)
```

```python
cur.execute("update tweets_sent_fa SET cleaned_text = remove_unnecessary(cleaned_text)")
conn.commit()
```

```python
op_keywords = {}
su_keywords = {}
```

```python
def process_tweets(row):
    for i in su_list01:
        if i in row:
            if i in su_keywords:
                su_keywords[i] += 1
            else:
                su_keywords[i] = 1
            return 'S'
    for i in op_list:
        if i in row:
            if i in op_keywords:
                op_keywords[i] += 1
            else:
                op_keywords[i] = 1
            return 'O'
    for i in su_list02:
        if i in row:
            if i in su_keywords:
                su_keywords[i] += 1
            else:
                su_keywords[i] = 1
            return 'S'
    return 'N'
```

```python
conn.create_function("process_tweets", 1, process_tweets)
```

```python
cur.execute("update tweets_sent_fa SET sentiment = process_tweets(cleaned_text) where sentiment is NULL")
conn.commit()
```
